```python
import sklearn
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
namesForlabels = data['target_names']
ourlabels = data['target']
namesForfeature = data['feature_names']
ourfeatures = data['data']
```

```python
print(namesForlabels)
```

```python
print(ourlabels)
```

```python
print(ourfeatures)
```

```python
from sklearn.model_selection import train_test_split

train, test, trainingLabels, testingLabels = train_test_split(ourfeatures, ourlabels, test_size = 0.33, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB

gaussiannb = GaussianNB()

ourModel = gaussiannb.fit(train, trainingLabels)
```

```python
ourPredictions = gaussiannb.predict(test)
print(ourPredictions)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(testingLabels, ourPredictions))
```

```python
#Creating a confusion matrix for the naive bayes classifier.
from sklearn import metrics 
from sklearn.naive_bayes import MultinomialNB # or any other NB model
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.model_selection import train_test_split

train, test, trainingLabels, testingLabels = train_test_split(ourfeatures, ourlabels, test_size = 0.33, random_state = 42)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
nb_classifier = MultinomialNB()
ourModel = gaussiannb.fit(train, trainingLabels)
nb_classifier.fit(train, trainingLabels)

ourlabels_pred = nb_classifier.predict(test)
class_names = [0,1]
acc_score = accuracy_score(testingLabels, ourlabels_pred)
#conf_mat = confusion_matrix(testingLabels, ourlabels_pred, labels = ['cat1', 'cat2', 'cat3', 'cat4', 'cat5'])
conf_mat = confusion_matrix(testingLabels, ourlabels_pred, labels = class_names)
print(acc_score)
print(conf_mat)

#setting up and visualizing the plot 
ax = plt.subplot()
tick_marks = np.arange(len(class_names)) 
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
sns.heatmap(pd.DataFrame(conf_mat), annot=True, cmap="YlGnBu" ,fmt='g') # Creating heatmap
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y = 1.1)
plt.ylabel('Actual diagnosis')
plt.xlabel('Predicted diagnosis')
```
